```python
import pandas as pd    # pandas is a data visualization library
import matplotlib.pyplot as plt      # matplotlib is for ploting the graph

# Step 1: Load the Hurricanes data from the CSV file and start visulizing the data with the imported packge up there

df = pd.read_csv('sample-salesv2.csv')

# Step 2: Inspect the data to understand its structure

print(df.head())  # This will print the first few rows of your dataset

# If you already have the DataFrame `df` loaded
mean_unit_price = df['unit price'].mean()

print("Mean Unit Price: ", mean_unit_price)
```

```python
# I have already loaded the DataFrame `df` and calculating the mean of the data
mean_unit_price = df['unit price'].mean()

mean_unit_price_2_decimals = round(mean_unit_price, 2) # thr round function adjusts the decimal place values

print("Mean Unit Price: ", mean_unit_price_2_decimals)
```

```python
# Calculate the median of the 'unit price' column
median_unit_price = df['unit price'].median()

# Print the median value with 2 decimal places
print(f"Median Unit Price: {median_unit_price:.2f}") # .2f set the two decimal place after the calculations
```

```python
# Calculate the standard deviation of the 'unit price' column
std_dev_unit_price = df['unit price'].std() # ".std" for the standared deviation

# Print the standard deviation with 2 decimal places
print(f"Standard Deviation of Unit Price: {std_dev_unit_price:.2f}")
```

```python
# Calculate the minimum and maximum of the 'unit price' column
min_unit_price = df['unit price'].min() # this is for the "min" values
max_unit_price = df['unit price'].max() # this is for the "max" values

# Print the results with 2 decimal places
print(f"Minimum Unit Price: {min_unit_price:.2f}") # it will give the 2 decimal place
print(f"Maximum Unit Price: {max_unit_price:.2f}") # it will give the 2 decimal place
```

```python
# Calculate the range of the 'unit price' column
range_unit_price = max_unit_price - min_unit_price # form the above calculated code

# Print the range with 2 decimal places
print(f"Range of Unit Price: {range_unit_price:.2f}") # giving the two decimal dataset
```

```python
# Calculate the 25th percentile (Q1) of the 'unit price' column
q1_unit_price = df['unit price'].quantile(0.25)

# Print the 25th percentile with 2 decimal places
print(f"25th Percentile (Q1) of Unit Price: {q1_unit_price:.2f}")
```

```python
# Calculate the 25th percentile (Q2) of the 'unit price' column
q2_unit_price = df['unit price'].quantile(0.50)

# Print the 25th percentile with 2 decimal places
print(f"25th Percentile (Q2) of Unit Price: {q2_unit_price:.2f}")
```

```python
# Calculate the 25th percentile (Q3) of the 'unit price' column
q3_unit_price = df['unit price'].quantile(0.75)

# Print the 25th percentile with 2 decimal places
print(f"25th Percentile (Q3) of Unit Price: {q3_unit_price:.2f}")
```

```python
# Calculate the Interquartile Range (IQR)
iqr_unit_price = q3_unit_price - q1_unit_price

# Print the IQR with 2 decimal places
print(f"Interquartile Range (IQR) of Unit Price: {iqr_unit_price:.2f}")
```

```python
# Calculate the lower and upper bounds
lower_bound = q1_unit_price - 1.5 * iqr_unit_price
upper_bound = q3_unit_price + 1.5 * iqr_unit_price

# Identify the outliers
outliers = df[(df['unit price'] < lower_bound) | (df['unit price'] > upper_bound)]

# Print out the outliers
print(f"Outliers in Unit Price: \n{outliers[['unit price']]}")
```

### the above output means, there is no outlires in the column of the dataset

```python
import matplotlib.pyplot as plt

# Define the data points
X = [1, 2, 3, 4]
Y = [20, 21, 20.5, 20.8]
y_error = [0.12, 0.13, 0.2, 0.1]  # Error for each y value ( it means the value can flactuate at the point of "y", eg:- it can flactuate from +,-0.12 or so on...

# Create a plot with customized figsize and dpi
plt.figure(figsize=(4, 5), dpi=100)  # Set figure size and dpi. dpi stands for "Dots per inch" and this works as pixel

# Create a plot with error bars
plt.errorbar(X, Y, 
             yerr=y_error,           # Error bars for the y-values
             fmt='-o',               # Line style and markers (o for circle markers)
             color='blue',           # Line color
             linestyle='-',          # Solid line
             markerfacecolor='yellow', # Marker fill color
             markeredgewidth=2,      # Marker border thickness
             markersize=8,           # Marker size
             label='Data with Error Bars', # Label for the legend
             capsize=5)              # Length of the caps on the error bars

# Add title and labels
plt.title('Simple Plot with Error Bars', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the DataFrame from the provided data
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'sex': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)

# Scatterplot of preTestScore vs postTestScore with the size of the points determined by age
plt.scatter(df['preTestScore'], df['postTestScore'], 
            s=df['age']*15,     # Size of each point is scaled by age
                                # s=df['age']*10: The size of each point (s) is scaled by the age column, 
                                # multiplied by 10 to make the size differences more noticeable.
            c=df['age'],        # Color each point based on age
            cmap='viridis',     # Colormap
            alpha=0.7)          # Transparency for better visibility

# Adding labels and title
plt.title('Scatter Plot of preTestScore vs postTestScore (Sized by Age)', fontsize=16)
plt.xlabel('preTestScore', fontsize=14)
plt.ylabel('postTestScore', fontsize=14)

# Show the color bar to indicate the color mapping based on age
plt.colorbar(label='Age') # this is for the colour bar label for the "age" column the right side of the plot

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the DataFrame from the provided data
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'sex': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)

# Scatter plot with size set to 300 and color determined by 'sex'
plt.scatter(df['preTestScore'], df['postTestScore'], 
            s=300,                      # Size of each point set to 300, "s" is for size
            c=df['sex'],                # Color determined by 'sex' (0 or 1), "c" is for colour
            cmap='plasma',            # Colormap for sex (different colors for males and females) "cmap" is to convert the numarical values to colours for ploting "PYiG, plasma etc." is the colour ploting code
            alpha=0.7)                  # Transparency for better visibility

# Adding labels and title
plt.title('Scatter Plot of preTestScore vs postTestScore (Sized 300, Colored by Sex)', fontsize=16)
plt.xlabel('preTestScore', fontsize=14)
plt.ylabel('postTestScore', fontsize=14)

# Show the color bar to indicate the color mapping based on sex
plt.colorbar(label='Sex (0=Male, 1=Female)') # as per the data male and female ploting

# Show the plot
plt.show()
```
